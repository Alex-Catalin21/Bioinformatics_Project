```python
dataset = pd.read_csv(r'C:\Users\Alex\Desktop\GitHub\Bioinformatics_Project\app\acetylcholinesterase_06_bioactivity_data_3class_pIC50_pubchem_fp.csv')
dataset
```

```python
# drop last column to assign it to y
X = dataset.drop(['pIC50'], axis=1)
X
```

```python
Y = dataset.iloc[:,-1]
Y
```

<h2><b>Remove low variance features</b></h2>

```python
from sklearn.feature_selection import VarianceThreshold

def remove_low_variance(input_data, threshold=0.1):
    selection = VarianceThreshold(threshold)
    selection.fit(input_data)
    return input_data[input_data.columns[selection.get_support(indices=True)]]

X = remove_low_variance(X, threshold=0.1)
X
```

```python
X.to_csv('descriptor_list.csv', index=False)
```

<h2><b>Random Forest Regression Model</b></h2>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=42)
model.fit(X, Y)
r2 = model.score(X, Y)
r2
```

<h2><b>Model Prediction</b></h2>

```python
Y_pred = model.predict(X)
Y_pred
```

<b><h2>Model Performance</h2></b>

```python
print('Mean squared error (MSE): %.2f'
     % mean_squared_error(Y, Y_pred))
print('Coefficient of determination (R^2): %.2f'
     % r2_score(Y, Y_pred))
```

<b><h2>Data Visualization(Experimental vs Predicted pIC50 for Training Data)</b></h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x=Y, y=Y_pred, c="#FFC680", alpha=0.3)

# Add trendline
# https://stackoverflow.com/questions/26447191/how-to-add-trendline-to-a-scatter-plot
z = np.polyfit(Y, Y_pred, 1)
p = np.poly1d(z)

plt.plot(Y,p(Y), "#C8A2C8")
plt.ylabel('Predicted pIC50')
plt.xlabel('Experimental pIC50')
```
